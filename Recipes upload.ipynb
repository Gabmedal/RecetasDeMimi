```python
from tkinter import *
import sqlite3

root = Tk()
root.title("Recipes upload")
root.geometry("600x600")

conectar = sqlite3.connect("Recipes.db")

c = conectar.cursor()

# create table

#c.execute("""CREATE TABLE Recetas_mimi1(
#    categoria text,
#   nombre text,
#   ingredientes text,
#   preparacion text
#   )""")

# create edit function to update record

def update():
        
    # Create a database to connect
    conectar = sqlite3.connect("Recipes.db")
    # Create a cursor
    c = conectar.cursor()
    
    registro_id = delete_box.get()
    c.execute("""UPDATE Recetas_mimi1 SET
        categoria = :categoria,
        nombre = :nombre,
        ingredientes = :ingredientes,
        preparacion = :preparacion
    
        WHERE oid = :oid""",
        {      
        "categoria": categoria_editor.get(),
        "nombre": nombre_editor.get(),
        "ingredientes": ingredientes_editor.get(),
        "preparacion": preparacion_editor.get(),
            
        "oid": registro_id
        })
    
     # commit changes
    conectar.commit()
    # close connection
    conectar.close()
    
    editor.destroy()

def myShow(): 
    myShow = ingredientes_editor.get(), preparacion_editor.get()
    myLabel = Label(editor, text=myShow)
    myLabel.grid(row=10, column=0)
   
def edit():
    global editor
    editor = Tk()
    editor.title("Actualizar receta")
    editor.geometry("400x400")
    
    # Create a database to connect
    conectar = sqlite3.connect("Recipes.db")
    # Create a cursor
    c = conectar.cursor()
    
    registro_id = delete_box.get()
    # create query database
    c.execute("SELECT * FROM Recetas_mimi1 WHERE oid = " + registro_id)
    registros = c.fetchall()
    
    # Create global variables for text box names
    global categoria_editor
    global nombre_editor
    global ingredientes_editor
    global preparacion_editor
    
    # Create text boxes
    categoria_editor = Entry(editor, width=30)
    categoria_editor.grid(row=0, column=1, padx=20, pady=(10,0))
    nombre_editor = Entry(editor, width=30)
    nombre_editor.grid(row=1, column=1)
    ingredientes_editor = Entry(editor, width=30)
    ingredientes_editor.grid(row=2, column=1)
    preparacion_editor = Entry(editor, width=30)
    preparacion_editor.grid(row=3, column=1)

    # Create text labels

    categoria_label = Label(editor, text="Categoria")
    categoria_label.grid(row=0, column=0, pady=(10,0))
    nombre_label = Label(editor, text="Nombre")
    nombre_label.grid(row=1, column=0)
    ingredientes_label = Label(editor, text="Ingredientes")
    ingredientes_label.grid(row=2, column=0)
    preparacion_label = Label(editor, text="Preparacion")
    preparacion_label.grid(row=3, column=0)
    
     #loop thru results
    for registro in registros:
        categoria_editor.insert(0, registro[0])
        nombre_editor.insert(0, registro[1])
        ingredientes_editor.insert(0, registro[2])
        preparacion_editor.insert(0, registro[3])
        
     #loop thru results
    print_registros = ""
    for registro in registros:
        print_registros += str(registro[2]) + "" + "\t" + str(registro[3]) + "" + "\n"
        
    # Create save button to save edited records
    save_btn = Button(editor, text="Guardar receta", command=update)
    save_btn.grid(row=8, column=0, columnspan=2, pady=10, padx=10, ipadx=100)
    
        # Create show button to show records
    mostrar_btn = Button(editor, text="Mostrar", command=myShow)
    mostrar_btn.grid(row=9, column=0, columnspan=2, pady=10, padx=10, ipadx=100)
    
    
# Create a function to delete record

def delete():
    conectar = sqlite3.connect("Recipes.db")
    c = conectar.cursor()
    
    #delete record
    c.execute("DELETE FROM Recetas_mimi1 WHERE oid= " + delete_box.get())
    delete_box.delete(0, END)
        
    # commit changes
    conectar.commit()
    # close connection
    conectar.close()


# create submit function

def submit():
    conectar = sqlite3.connect("Recipes.db")
    c = conectar.cursor()
    
    #insert into table
    c.execute("INSERT INTO Recetas_mimi1 VALUES (:categoria, :nombre, :ingredientes, :preparacion)", 
            {
              "categoria": categoria.get(),
              "nombre": nombre.get(),
              "ingredientes": ingredientes.get(),
              "preparacion": preparacion.get()
            })
    
    
    #clear text boxes
    categoria.delete(0, END)
    nombre.delete(0, END)
    ingredientes.delete(0, END)
    preparacion.delete(0, END)
     
    # commit changes
    conectar.commit()
    # close connection
    conectar.close()

# create query funcion

def query():
    conectar = sqlite3.connect("Recipes.db")
    c = conectar.cursor()
    
    # query database
    c.execute("SELECT *, oid FROM Recetas_mimi1")
    registros = c.fetchall()
    # print(registros)

    # fetchmany(50), fetchone, etc
    
    #loop thru results
    print_registros = ""
    for registro in registros:
        print_registros += str(registro[1]) + "" + "\t" + str(registro[4]) + "" + "\n"
        
    query_label = Label(root, text=print_registros)
    query_label.grid(row=13, column=0, columnspan=2)
    
    # commit changes
    conectar.commit()
    # close connection
    conectar.close()
    
# Create text boxes
categoria = Entry(root, width=40)
categoria.grid(row=0, column=1, padx=20, pady=(10,0))
nombre = Entry(root, width=40)
nombre.grid(row=1, column=1)
ingredientes = Entry(root, width=40)
ingredientes.grid(row=2, column=1)
preparacion = Entry(root, width=40)
preparacion.grid(row=3, column=1)
delete_box = Entry(root, width=40)
delete_box.grid(row=10, column=1, pady=5)

# Create text labels

categoria_label = Label(root, text="Categoria")
categoria_label.grid(row=0, column=0, pady=(10,0))
nombre_label = Label(root, text="Nombre")
nombre_label.grid(row=1, column=0)
ingredientes_label = Label(root, text="Ingredientes")
ingredientes_label.grid(row=2, column=0)
preparacion_label = Label(root, text="Preparacion")
preparacion_label.grid(row=3, column=0)
delete_box_label = Label(root, text="Seleccione receta")
delete_box_label.grid(row=10, column=0, pady=5)

# Create submit button

submit_btn = Button(root, text="Agregar a lista de recetas", command=submit)
submit_btn.grid(row=7, column=0, columnspan=2, pady=10, padx=10, ipadx=115)

#  create a query button
query_btn = Button(root, text="Mostrar recetas", command=query)
query_btn.grid(row=8, column=0, columnspan=2, pady=10, padx=10, ipadx=137) 

# Create Delete button
delete_btn = Button(root, text="Eliminar receta", command=delete)
delete_btn.grid(row=11, column=0, columnspan=2, pady=10, padx=10, ipadx=137)

# Create update button
edit_btn = Button(root, text="Actualizar receta", command=edit)
edit_btn.grid(row=12, column=0, columnspan=2, pady=10, padx=10, ipadx=137)

# commit changes
conectar.commit()

# close connection
conectar.close()

root.mainloop()
```
